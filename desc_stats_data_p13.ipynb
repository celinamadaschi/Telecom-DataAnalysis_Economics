```python
import pandas as pd

data = pd.read_csv('repec_with_gender.csv')
data.info()
```

```python
# descriptive statistics of the numerical data

data.describe()
```

```python
data.head()
```

```python
# dans la colonne 'gender' remplacer male par 0 et female par 1 et unknown par NaN

import numpy as np

# Définition du dictionnaire de mapping
mapping = {
    'male': 0,
    'female': 1,
    'unknown': np.nan
}

# Application du mapping sur la colonne 'gender'
data['gender'] = data['gender'].map(mapping)
```

```python
# faire de meme pour type

mapping = {
    'Journal Article': 1,
    'Working Paper': 0,
    'unknown': np.nan
}

# Application du mapping sur la colonne 'gender'
data['Type'] = data['Type'].map(mapping)
```

```python
data.drop_duplicates(inplace=True)
```
