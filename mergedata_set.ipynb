```python
import re
import unidecode

def clean(s):
    if not isinstance(s, str):
        return ""
    s = unidecode.unidecode(s.lower())
    s = re.sub(r"[^a-z\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s
```

```python
df1 = df1.copy()

df1["last_name_clean"] = (
    df1["author_name"]
    .str.split(",", n=1)
    .str[0]
    .apply(clean)
)

df1["first_names_clean"] = (
    df1["author_name"]
    .str.split(",", n=1)
    .str[1]
    .apply(clean)
)

```

```python
df1["tokens_df1"] = df1.apply(
    lambda r: set(
        (r["first_names_clean"] + " " + r["last_name_clean"]).split()
    ),
    axis=1
)
```

```python
def drop_initials(tokens):
    return {t for t in tokens if len(t) > 1}

df1["tokens_df1"] = df1["tokens_df1"].apply(drop_initials)
```

```python
df1["ambiguous_name"] = df1["tokens_df1"].apply(lambda x: len(x) <= 1)
```

```python
df1["tokens_df1"] = df1.apply(
    lambda r: set(
        (r["first_names_clean"] + " " + r["last_name_clean"]).split()
    ),
    axis=1
)

df1["tokens_df1"] = df1["tokens_df1"].apply(drop_initials)

```

```python
df2_auth = (
    df2
    .dropna(subset=["Author(s)"])
    .assign(author_raw=df2["Author(s)"].str.split(";"))
    .explode("author_raw")
)

df2_auth["author_raw"] = df2_auth["author_raw"].str.strip()
```

```python
df2_auth["author_clean"] = df2_auth["author_raw"].apply(clean)

df2_auth["tokens_df2"] = (
    df2_auth["author_clean"]
    .str.split()
    .apply(set)
    .apply(drop_initials)
)
```

```python
def is_match(tokens1, tokens2):
    return (
        isinstance(tokens1, set)
        and isinstance(tokens2, set)
        and len(tokens1) > 0
        and tokens1.issubset(tokens2)
    )
```
