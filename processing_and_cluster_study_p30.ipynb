```python
import pandas as pd

data = pd.read_csv('final_author_data_JEL_2010_full.csv')


data.head(10)
```

```python
# renommer Author(s) en Publication Info, Year en Author(s) et Publication Info en Year
data = data.rename(columns={'Author(s)': 'Publication Info', 'Year': 'Author(s)', 'Publication Info': 'Year'})
data.head(10)
```

```python
# on split URL en Type et Journal au niveau de la premiere virgule
split_df = data['URL'].str.split(r',\s*', n=1, expand=True)
split_df.columns = ['Type', 'Journal']
data = data.join(split_df)
data.head(2)
```

```python
#enlever les colonnes 1 et 3
data = data.drop(data.columns[[2]], axis=1)
data.head(1)
```

```python
#remplacer les valeurs manquantes par 0
data = data.fillna(0)

# mettre Year en int
data['Year'] = data['Year'].astype(int)
data.head(1)
```

```python
# ajouter une colonne 'collaboration' avec une liste vide 
data['collaboration'] = [[] for _ in range(len(data))]
data.head(1)
```

```python
# 1. Nettoyage des NaN
data = data.dropna(subset=['Author(s)'])

# 2. On transforme la chaîne en une vraie liste Python
# (Plus facile à manipuler que des colonnes séparées Author_1, Author_2...)
data['authors_list'] = data['Author(s)'].str.split(r'\s*&\s*')

# 3. On extrait le Premier Auteur (l'index 0 de la liste)
data['Main_Author'] = data['authors_list'].str[0]

# 4. On met TOUS les autres dans 'collaboration' (slicing de liste)
# La syntaxe [1:] veut dire "de l'élément 1 jusqu'à la fin"
data['collaboration'] = data['authors_list'].apply(lambda x: x[1:] if isinstance(x, list) else [])

# 5. Sélection finale des colonnes
data = data[['Source Author Profile','Title', 'Main_Author', 'collaboration', 'Year','JEL_Codes_Extracted', 'Type','Journal']]

# Affichag
data.head(2)
```

```python
# enlever les virgules dans les noms des auteurs
data['Main_Author'] = data['Main_Author'].str.replace(',', '')
data.head(10)
```

```python
# afficher la premiere occurence de la colonne 'Journal'

data['Journal'].iloc[0]
```

```python
# enlever la colonne 'Journal'
data = data.drop(columns=['Journal'])
```

```python
data_a_z = pd.read_csv('data_A_Z.csv')

data_a_z.head(10)
```

```python
# merge data_a_z avec author_stats sur les colonnes Main_Author
# garder uniquement autor_url de data_a_z et institution de data_a_z
data_a_z = data_a_z[['author_url', 'institution']]

data = data.rename(columns={'Source Author Profile': 'author_url'})
merged_data = pd.merge(data, data_a_z, on='author_url', how='left')
merged_data.head(10)
```

```python
import pandas as pd
import ast

author_directory = merged_data[['Main_Author', 'institution']].dropna().drop_duplicates('Main_Author', keep='last')
author_map = pd.Series(author_directory.institution.values, index=author_directory.Main_Author).to_dict()

print(f"Annuaire créé avec {len(author_map)} auteurs et leurs institutions.")

def parse_list(val):
    if isinstance(val, str):
        try:
            return ast.literal_eval(val)
        except:
            return []
    return val if isinstance(val, list) else []

merged_data['collaboration'] = merged_data['collaboration'].apply(parse_list)

def get_full_institution_list(row):
    institutions = [row['institution']]
    
    co_authors = row['collaboration']
    
    for author_name in co_authors:
        if author_name in author_map:
            institutions.append(author_map[author_name])
        else:
            pass 
            
    return list(set([inst for inst in institutions if inst]))

merged_data['reconstructed_affiliations'] = merged_data.apply(get_full_institution_list, axis=1)

print(merged_data[['Main_Author', 'institution', 'collaboration', 'reconstructed_affiliations']].head())
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import friedmanchisquare
import itertools

period_df = merged_data[(merged_data['Year'] >= 2015) & (merged_data['Year'] <= 2024)].copy()

G = nx.Graph()
for affils in period_df['reconstructed_affiliations']:
    unique = sorted(list(set(affils)))
    if len(unique) > 1:
        edges = list(itertools.combinations(unique, 2))
        for u, v in edges:
            if G.has_edge(u, v): G[u][v]['weight'] += 1
            else: G.add_edge(u, v, weight=1)

communities = nx.community.louvain_communities(G, weight='weight')
top_5_communities = sorted(communities, key=len, reverse=True)[:5]

cluster_defs = {}
cluster_names = []
for idx, comm in enumerate(top_5_communities):
    subgraph = G.subgraph(comm)
    leader = max(dict(subgraph.degree(weight='weight')), key=dict(subgraph.degree(weight='weight')).get)
    name = f"{leader} & Co"
    cluster_defs[name] = set(comm)
    cluster_names.append(name)

rank_matrix = []
years = sorted(period_df['Year'].unique())

for year in years:
    year_df = period_df[period_df['Year'] == year]
    volumes = []
    for name in cluster_names:
        count = year_df[year_df['reconstructed_affiliations'].apply(lambda x: any(i in cluster_defs[name] for i in x))].shape[0]
        volumes.append(count)
    ranks = pd.Series(volumes).rank(ascending=False, method='min').tolist()
    rank_matrix.append(ranks)

#Test statistique
data_transposed = list(map(list, zip(*rank_matrix))) 
stat, p_value = friedmanchisquare(*data_transposed)

# W de Kendall (0 à 1)
N, k = len(years), len(cluster_names)
kendall_w = stat / (N * (k - 1))

print(f"--- RÉSULTATS STABILITÉ DU CLASSEMENT ---")
print(f"P-value       : {p_value:.5f}")
print(f"W de Kendall  : {kendall_w:.4f}")

if p_value < 0.05:
    print("Classement significativement stable")
    if kendall_w > 0.7: print("-> Hiérarchie très verrouillée (W > 0.7)")
    elif kendall_w > 0.3: print("-> Hiérarchie modérée")
    else: print("-> Accord faible")
else:
    print("Significatif ? : Classement instable / Aléatoire.")
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import itertools

period_df = merged_data[(merged_data['Year'] >= 2015) & (merged_data['Year'] <= 2024)].copy()

G = nx.Graph()
for affils in period_df['reconstructed_affiliations']:
    unique = sorted(list(set(affils)))
    if len(unique) > 1:
        edges = list(itertools.combinations(unique, 2))
        for u, v in edges:
            if G.has_edge(u, v): G[u][v]['weight'] += 1
            else: G.add_edge(u, v, weight=1)

communities = nx.community.louvain_communities(G, weight='weight')

node_to_cluster = {}
for idx, comm in enumerate(communities):
    for node in comm:
        node_to_cluster[node] = idx

trends_data = []
years = sorted(period_df['Year'].unique())

for year in years:
    year_df = period_df[period_df['Year'] == year]
    collab_papers = year_df[year_df['collaboration'].apply(len) > 0]
    
    if len(collab_papers) < 10: continue
    
    total = 0
    silo_inst_count = 0  
    silo_cluster_count = 0 
    
    for affils in collab_papers['reconstructed_affiliations']:
        if len(affils) == 0: continue
        total += 1
        
        unique_insts = set(affils)
        if len(unique_insts) == 1:
            silo_inst_count += 1
            
        cluster_ids = {node_to_cluster[inst] for inst in affils if inst in node_to_cluster}
        if len(cluster_ids) == 1:
            silo_cluster_count += 1
            
    if total > 0:
        trends_data.append({
            'Year': year,
            'Silo_Institution': (silo_inst_count / total) * 100,
            'Silo_Cluster': (silo_cluster_count / total) * 100
        })

df_trends = pd.DataFrame(trends_data)

slope1, _, _, p1, _ = stats.linregress(df_trends['Year'], df_trends['Silo_Institution'])
slope2, _, _, p2, _ = stats.linregress(df_trends['Year'], df_trends['Silo_Cluster'])

print(f"{'INDICATOR':<25} | {'SLOPE':<8} | {'P-VALUE':<8} | {'VERDICT'}")
print("-" * 70)
for label, slope, p_val in [('Intra-Institution', slope1, p1), ('Intra-Cluster', slope2, p2)]:
    sig = "***" if p_val < 0.001 else "**" if p_val < 0.01 else "*" if p_val < 0.05 else "ns"
    print(f"{label:<25} | {slope:+.3f}   | {p_val:.4f}   | {sig}")
print("-" * 70)

sig1 = "***" if p1 < 0.001 else "**" if p1 < 0.01 else "*" if p1 < 0.05 else "ns"
sig2 = "***" if p2 < 0.001 else "**" if p2 < 0.01 else "*" if p2 < 0.05 else "ns"

fig, ax1 = plt.subplots(figsize=(12, 7))

color1 = 'tab:blue'

sns.lineplot(data=df_trends, x='Year', y='Silo_Institution', ax=ax1, 
             marker='o', color=color1, linewidth=3, label='Institution')
sns.regplot(data=df_trends, x='Year', y='Silo_Institution', ax=ax1, scatter=False, 
            color=color1, line_kws={'linestyle':'--', 'alpha':0.5})

ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('% Intra-Institution', color=color1, fontsize=13, fontweight='bold')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(True, axis='x', alpha=0.3) 

ax2 = ax1.twinx()
color2 = 'tab:red'

sns.lineplot(data=df_trends, x='Year', y='Silo_Cluster', ax=ax2, 
             marker='s', color=color2, linewidth=3, label='Cluster')
sns.regplot(data=df_trends, x='Year', y='Silo_Cluster', ax=ax2, scatter=False, 
            color=color2, line_kws={'linestyle':'--', 'alpha':0.5})

ax2.set_ylabel('% Intra-Cluster', color=color2, fontsize=13, fontweight='bold')
ax2.tick_params(axis='y', labelcolor=color2)

ax1.set_xticks(df_trends['Year'])
plt.tight_layout()
plt.show()
```
